# Exploratory Data Analysis (EDA)

📊 Analysis 1: Department-wise Attrition
### Business Question:
Which department has the highest employee attrition?





```python
department_attrition = df[df["Attrition"] == "Yes"]["Department"].value_counts()

department_attrition
```

```python
import matplotlib.pyplot as plt

department_attrition.plot(kind="bar", figsize=(8,5))

plt.title("Department-wise Employee Attrition")
plt.xlabel("Department")
plt.ylabel("Employees Left")

plt.show()
```

### 📌 Business Insight

- Research & Development department has the highest employee attrition.
- Sales department has the second highest attrition.
- Human Resources has the lowest attrition.
- HR should focus on improving employee retention in the Research & Development and Sales departments.

# 📊 Analysis 2: Overtime vs Attrition

## 🎯 Business Question
Does working overtime increase employee attrition?

```python
# Create a cross-tabulation of OverTime and Attrition
overtime_attrition = pd.crosstab(df["OverTime"], df["Attrition"])

overtime_attrition
```

```python
plt.figure(figsize=(7,5))

overtime_attrition.plot(kind="bar")

plt.title("OverTime vs Attrition")

plt.xlabel("OverTime")

plt.ylabel("Number of Employees")

plt.xticks(rotation=0)

plt.show()
```

## 📌 Business Insight

- Employees who work overtime have a significantly higher attrition rate than employees who do not work overtime.
- Most employees who do not work overtime remain with the company.
- Overtime appears to be one of the major factors contributing to employee attrition.
- HR should review overtime policies and promote a better work-life balance to improve employee retention.

## 💼 HR Recommendation

- Reduce unnecessary overtime.
- Monitor employees who consistently work overtime.
- Introduce flexible work schedules.
- Encourage managers to distribute workload more effectively.

# 📊 Analysis 3: Gender-wise Attrition

## 🎯 Business Question

Is employee attrition different between male and female employees?

```python
gender_attrition = df[df["Attrition"]=="Yes"]["Gender"].value_counts()

gender_attrition
```

```python
plt.figure(figsize=(6,4))

plt.bar(gender_attrition.index,
        gender_attrition.values)

plt.title("Gender-wise Employee Attrition")

plt.xlabel("Gender")

plt.ylabel("Employees Left")

plt.show()
```

## 📌 Business Insight

- More male employees left the company than female employees in terms of total count.
- Female employees had fewer resignations than male employees.
- This analysis is based on employee counts only. To compare attrition fairly, attrition rates should also be calculated for each gender.

## 💼 HR Recommendation

- Conduct employee satisfaction surveys for both male and female employees.
- Analyze additional factors such as overtime, salary, and job role to understand the root causes of attrition.
- Focus retention strategies on high-risk employee groups identified through further analysis.

# 📊 Analysis 4: Age-wise Attrition

## 🎯 Business Question

Which age groups experience the highest employee attrition?

```python
age_attrition = df[df["Attrition"]=="Yes"]["Age"]

plt.figure(figsize=(10,5))

plt.hist(age_attrition, bins=10)

plt.title("Age Distribution of Employees Who Left")

plt.xlabel("Age")

plt.ylabel("Number of Employees")

plt.show()
```

## 📌 Business Insight

- Most employees who left the company were between **26 and 35 years** of age.
- Employee attrition gradually decreases as age increases.
- Younger employees appear to be more likely to leave the organization than older employees.
- This may indicate that younger employees are more likely to seek better career opportunities, higher salaries, or faster career growth.

## 💼 HR Recommendation

- Develop career growth programs for young employees.
- Introduce mentorship and training opportunities.
- Review compensation and promotion policies for early-career employees.
- Conduct regular employee engagement surveys to understand the needs of younger employees.

# 📊 Analysis 5: Age Group-wise Attrition

## 🎯 Business Question

Which age group has the highest employee attrition?

```python
# Create Age Groups
bins = [18,25,35,45,55,65]

labels = ['18-25','26-35','36-45','46-55','56+']

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
age_group_attrition = df[df["Attrition"]=="Yes"]["AgeGroup"].value_counts().sort_index()

age_group_attrition
```

```python
plt.figure(figsize=(8,5))

plt.bar(age_group_attrition.index.astype(str),
        age_group_attrition.values)

plt.title("Age Group-wise Employee Attrition")

plt.xlabel("Age Group")

plt.ylabel("Employees Left")

plt.show()
```

# 🎂 Age Distribution by Attrition

This visualization compares employee age with attrition to identify the age groups most likely to leave the company.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='Age',
    hue='Attrition',
    multiple='stack',
    bins=20
)

plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Employee Count')

plt.show()
```

# 🏢 Department-wise Attrition

This analysis identifies which departments experience the highest employee attrition.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df[df['Attrition']=='Yes'],
    x='Department',
    order=df[df['Attrition']=='Yes']['Department'].value_counts().index
)

plt.title('Department-wise Employee Attrition')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.xticks(rotation=15)

plt.show()
```

# ⏰ Overtime vs Attrition

This analysis examines whether employees working overtime are more likely to leave the organization.

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='OverTime',
    hue='Attrition'
)

plt.title('Overtime vs Attrition')
plt.xlabel('OverTime')
plt.ylabel('Employee Count')

plt.show()
```

# 💰 Monthly Income vs Attrition

This box plot compares the monthly income of employees who stayed with those who left the organization.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Attrition',
    y='MonthlyIncome'
)

plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')

plt.show()
```

# 📅 Years at Company vs Attrition

This analysis evaluates whether employees with fewer years at the company are more likely to resign.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Attrition',
    y='YearsAtCompany'
)

plt.title('Years at Company vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')

plt.show()
```

# 😊 Job Satisfaction vs Attrition

This visualization analyzes the relationship between employee job satisfaction and attrition.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x='JobSatisfaction',
    hue='Attrition'
)

plt.title('Job Satisfaction vs Attrition')
plt.xlabel('Job Satisfaction (1 = Low, 4 = High)')
plt.ylabel('Employee Count')

plt.show()
```

# 🔥 Correlation Heatmap

The correlation heatmap illustrates relationships between numerical features in the dataset. Stronger positive correlations are shown with warmer colors, while weaker relationships appear cooler.

```python
plt.figure(figsize=(18,12))

numeric_df = df.select_dtypes(include='number')

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm',
    annot=False
)

plt.title('Correlation Heatmap')

plt.show()
```

# 📌 Key Findings

The analysis reveals several important insights:

- Laboratory Technicians have the highest employee attrition.
- Employees aged between **25–35 years** show the highest turnover.
- Research & Development has the highest number of employee exits.
- Employees working overtime are significantly more likely to leave.
- Lower monthly income is associated with higher attrition.
- Employees with fewer years at the company tend to resign more frequently.
- Lower job satisfaction is linked with increased employee attrition.

# ✅ Conclusion

This project analyzed the IBM HR Analytics Employee Attrition dataset to identify factors contributing to employee turnover.

The analysis indicates that **overtime, lower salary, younger age, shorter tenure, and lower job satisfaction** are the major factors associated with employee attrition.
